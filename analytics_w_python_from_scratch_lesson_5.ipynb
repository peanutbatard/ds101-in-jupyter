### 3. Magic trick: recursion

A recursive function is a function that calls itself

```python
def not_bright():
    print('this is so dumb')
    not_bright()
```

```python
not_bright()
```

```python
def f(x, n):
    if n == 0:
        return 1
    else:
        return x * f(x, n - 1)
```

```python
f(2, 3)
```

f(2, 3) = 2 * f(2, 2)

f(2, 2) = 2 * f(2, 1)

f(2, 1) = 2 * f(2, 0)

f(2, 0) = 1

### 5. Analytics example: median calculation

```python
a = [83, 27, 45, 56, 98, 21, 10, 87, 10, 22, 6]
b = [10, 76, 24, 12]
c = [20, 30]
```

Median for a

```python
my_a = a.copy()
my_a.sort()
my_a
```

```python
len(my_a)
```

```python
index = len(my_a) // 2
```

```python
my_a[index]
```

Median for b

```python
my_b = b.copy()
my_b.sort()
index = len(my_b) // 2
my_b
```

```python
my_b[index]
```

```python
(my_b[index] + my_b[index - 1]) / 2
```

Let's build a function

```python
def my_med(nums):
    my_nums = nums.copy()
    my_nums.sort()
    length = len(my_nums)
    index = length // 2
    if length % 2 != 0: # off number of elements
        return my_nums[index]
    else: # even number of elements
        return (my_nums[index] + my_nums[index - 1]) / 2
```

```python
my_med(a)
```

```python
my_med(b)
```

```python
my_med(c)
```
